```python
import pandas as pd
df=pd.read_csv(r"D:\Road to Data Science\DataSets Given In Class\housing.csv.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
numeric_lst=['int64','float64']
cols=list(df.select_dtypes(include=numeric_lst).columns)
```

```python
df=df[cols]
df
```

```python
X=df.drop(['SalePrice'],axis=1)
y=df['SalePrice']
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=0)
X_train.isna().sum()
```

```python
from sklearn.feature_selection import mutual_info_regression
mutual_info=mutual_info_regression(X_train.fillna(0),y_train)
mutual_info
```

```python
mutual_info=pd.Series(mutual_info)
mutual_info.index=X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(10,6))
```

```python
from sklearn.feature_selection import SelectPercentile
sel_percentile=SelectPercentile(mutual_info_regression,percentile=20)
sel_percentile.fit(X_train.fillna(0),y_train)
X_train.columns[sel_percentile.get_support()]
```
